# Global Terrorism  Dataset Analysis

```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df=pd.read_csv('globalterrorismdb_0718dist.csv',encoding='ISO-8859-1')
```

```python
print(df)
```

### 1. Number of Terrorist Incidents Across Regions from 1970-2017

```python
df1=df.groupby(['region_txt'])[['eventid']].count()
```

```python
print(df1)
```

```python
df1=df1.reset_index()
df1=df1.sort_values('eventid',ascending=False)
```

```python
bar=px.bar(x=df1['region_txt'],y=df1['eventid'],color=df1['region_txt'],title='Number of Terrorist Incidents Across Regions from 1970-2017',labels={
    'x':'Region',
    'y':'Number of Terrorist Incidents',
    'color':'Region'
})
bar.show()
```

### 2. Total Number of Terrorist Incidents By Countries

```python
data_coun=df.groupby('country_txt')['eventid'].count()
data_coun=data_coun.reset_index()
```

```python
fig=go.Figure(data=go.Choropleth(locations=data_coun['country_txt'],locationmode='country names',z=data_coun['eventid'],text = data_coun['country_txt'],colorscale = 'plasma', colorbar_title = 'Number of Terrorists Incidents'))

fig.update_layout(title_text = 'Total Number of Terrorists Incidents by Countries',geo = dict(showframe = False,showcoastlines = True,projection_type = 'equirectangular'),)

fig.show()
```

### 3. Top 10 Countries with Most Number of Terrorists Incidents From 1970-2017

```python
dt_coun=data_coun.sort_values('eventid',ascending=False)
dt_coun=dt_coun[0:10]
```

```python
bar=px.bar(x=dt_coun['country_txt'],y=dt_coun['eventid'],color=dt_coun['country_txt'], title='Top 10 Countries with Most Number of Terrorists Incidents From 1970-2017',
          labels={
              'x':'Countries',
              'y':'Number of Terrorist Incidents',
              'color':'Countries'
          })
bar.show()
```
